```python
import os

# Folder containing your stop word files
folder_path = 'F:/Blackcoffer assignment/StopWords-20230812T205218Z-001/StopWords'

# Create a list to store the paths of stop word files
stopword_files = []

# Iterate over the files in the folder
for filename in os.listdir('F:/Blackcoffer assignment/StopWords-20230812T205218Z-001/StopWords'):
    # Check if the file is a text file
    if filename.endswith('.txt'):
        # Append the full path of the file to the list
        stopword_files.append(os.path.join('F:/Blackcoffer assignment/StopWords-20230812T205218Z-001/StopWords', filename))

# Now stopword_files will contain the paths of all text files in the folder
# You can use this list in your code to open each file and read its contents
print(stopword_files)
```

```python
# Define the path of the new file
new_file_path = 'F:/Blackcoffer assignment/StopWords-20230812T205218Z-001/StopWords/AllStopWords.txt'

# Open the new file in write mode
with open(new_file_path, 'w') as new_file:
    # Iterate through each file in the stopword_files list
    for file_path in stopword_files:
        # Open each file in read mode
        with open(file_path, 'r') as file:
            # Read the contents of the file
            content = file.read()
            # Write the contents to the new file
            new_file.write(content)
            # Add a newline character to separate contents from different files
            new_file.write('\n')
```
